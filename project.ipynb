```python
import numpy as np # Image Processing 
import fitz ,os,glob,cv2,pathlib  # Pdf To Image 
from PIL import Image # For Reading Image (1)
from matplotlib.image import imread # For Reading Image (2)
from sklearn.model_selection import train_test_split #Dividing Data 
from sklearn.ensemble import RandomForestClassifier #ML MODEL
```

```python
from keras.preprocessing.image import img_to_array,load_img,ImageDataGenerator # For Reading Image (3)
```

```python
def extract_image(file,name):   #Convert Document To Image 
    pdf = fitz.open(file) #Read File 
    page = pdf.loadPage(0) # Get Page 0 
    pix = page.get_pixmap() # Convert To Pixels 
    pix.save("images/"+ name +".jpg") #Save pixels to Image 
    print(f'image {file} Saved')
```

```python
files = glob.glob("files/*.pdf") #Getting ALL Files 
```

```python
files[0]
```

```python
for i in range(len(files)):
    extract_image(files[i],name="file" + str(i))
```
